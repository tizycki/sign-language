```python
import glob
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
import json
```

```python
dir_path = './data/video_frames/internal_resource_2337_fps10/'
image_extension = 'jpg'
```

```python
def image_gen(start_idx=0):
    files = np.sort(glob.glob(f'{dir_path}*.{image_extension}'))
    for idx, file in enumerate(files):
        if idx < start_idx:
            continue
            
        image = cv2.imread(file)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        yield image, file
```

```python
labeling_output = []

for results in image_gen():
    file = results[1]
    img = results[0]

    print(os.path.basename(file))
    print(os.path.basename(file).split('_')[-1])
    
    fig = plt.figure(figsize=(16,9))
    plt.imshow(img)
    plt.show()
    
    flag = input('Blank for stop_pose=0:')
    
    clear_output(wait=True)
    
    if flag == 'break':
        break
    
    if flag != '':
        labeling_output.append({
            'file_path': os.path.realpath(file),
            'stop_pose': 1
        })
    else:
        labeling_output.append({
            'file_path': os.path.realpath(file),
            'stop_pose': 0
        })

t = time.localtime()
timestamp = time.strftime('%b-%d-%Y_%H%M', t)

output_file_path = os.path.join('./data', f'stop_pose_labels{timestamp}.json')
with open(output_file_path, 'w') as outfile:
    json.dump(labeling_output, outfile)
```
